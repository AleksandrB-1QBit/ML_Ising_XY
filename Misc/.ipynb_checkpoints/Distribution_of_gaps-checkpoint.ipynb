```python
import numpy as np
import qit
import matplotlib.pyplot as plt
```

```python
# Define 1-qubit Pauli operators

I = qit.lmap(np.array([[1, 0], [0, 1]]))
X = qit.lmap(np.array([[0, 1], [1, 0]]))
Y = qit.lmap(np.array([[0, -1j], [1j, 0]]))
Z = qit.lmap(np.array([[1, 0], [0, -1]]))
```

```python
# Define 10-qubit Pauli operators

X_1  = qit.tensor(X,I,I,I,I,I,I,I,I,I)
X_2  = qit.tensor(I,X,I,I,I,I,I,I,I,I)
X_3  = qit.tensor(I,I,X,I,I,I,I,I,I,I)
X_4  = qit.tensor(I,I,I,X,I,I,I,I,I,I)
X_5  = qit.tensor(I,I,I,I,X,I,I,I,I,I)
X_6  = qit.tensor(I,I,I,I,I,X,I,I,I,I)
X_7  = qit.tensor(I,I,I,I,I,I,X,I,I,I)
X_8  = qit.tensor(I,I,I,I,I,I,I,X,I,I)
X_9  = qit.tensor(I,I,I,I,I,I,I,I,X,I)
X_10 = qit.tensor(I,I,I,I,I,I,I,I,I,X)

Z_1  = qit.tensor(Z,I,I,I,I,I,I,I,I,I)
Z_2  = qit.tensor(I,Z,I,I,I,I,I,I,I,I)
Z_3  = qit.tensor(I,I,Z,I,I,I,I,I,I,I)
Z_4  = qit.tensor(I,I,I,Z,I,I,I,I,I,I)
Z_5  = qit.tensor(I,I,I,I,Z,I,I,I,I,I)
Z_6  = qit.tensor(I,I,I,I,I,Z,I,I,I,I)
Z_7  = qit.tensor(I,I,I,I,I,I,Z,I,I,I)
Z_8  = qit.tensor(I,I,I,I,I,I,I,Z,I,I)
Z_9  = qit.tensor(I,I,I,I,I,I,I,I,Z,I)
Z_10 = qit.tensor(I,I,I,I,I,I,I,I,I,Z)
```

```python
# Define 1-D quantum Ising Hamiltonian with both transverse and longitudinal fields on an open chain
# .data represents tensor as np.array

def H(t, l):
    tmp_1  = l * (Z_1.data + Z_2.data + Z_3.data + Z_4.data + Z_5.data + Z_6.data 
                  + Z_7.data + Z_8.data + Z_9.data + Z_10.data)
    tmp_2  = t * (X_1.data + X_2.data + X_3.data + X_4.data + X_5.data + X_6.data 
                  + X_7.data + X_8.data + X_9.data + X_10.data)
    tmp_3  = Z_1.data  @ Z_2.data
    tmp_4  = Z_2.data  @ Z_3.data
    tmp_5  = Z_3.data  @ Z_4.data
    tmp_6  = Z_4.data  @ Z_5.data
    tmp_7  = Z_5.data  @ Z_6.data
    tmp_8  = Z_6.data  @ Z_7.data
    tmp_9  = Z_7.data  @ Z_8.data
    tmp_10 = Z_8.data  @ Z_9.data
    tmp_11 = Z_9.data  @ Z_10.data
    return tmp_1 + tmp_2 + tmp_3 + tmp_4 + tmp_5 + tmp_6 + tmp_7 + tmp_8 + tmp_9 + tmp_10 + tmp_11
```

```python
t = 0.5
l = np.linspace(0,0.1,21)

w = np.zeros((len(l), 2 ** 10))
v = np.zeros((len(l), 2 ** 10, 2 ** 10))
```

```python
%%time
# Diagonalize the model
for j in range(len(l)):
    w[j,:], v[j,:,:] = np.linalg.eigh(H(t,l[j]))
```

```python
# Calculate energy gaps for every l
spacings = np.zeros((len(l), 2 ** 10))
for i in range(len(l)):
    for j in range(2 ** 10 -1):
        spacings[i,j] = w[i,j+1] - w[i,j]
```

```python
# See how distribution changes from Poisson to Wigner when we move to an unintegrable(chaotic) mode
for i in range(len(l)):
    plt.hist(spacings[i,:], bins=1024, range=[0,0.05])
    plt.show()
```
