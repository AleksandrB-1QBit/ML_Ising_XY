```python
import numpy as np
from scipy.sparse import csr_matrix, kron, eye
import scipy.sparse.linalg as las
import scipy.linalg as la
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
I = csr_matrix(np.array([[1, 0], 
                         [0, 1]]))
X = csr_matrix(np.array([[0, 1], 
                         [1, 0]]))
Y = csr_matrix(np.array([[0, -1j], 
                         [1j, 0]]))
Z = csr_matrix(np.array([[1, 0], 
                         [0, -1]]))
```

Now let's make a (recursive) function that will build the Hamiltonian for us

$$H = \sum_{i=1}^{L-1} Z_i\,Z_{i+1} + \sum_{i=1}^L (\tau\,X_i+\lambda\,Z_i)$$

```python
def Ham(L=3, ta = 0., la = 0.):
    if L == 2:
        return kron(Z,Z)+ta*(kron(X,I)+kron(I,X))+la*(kron(Z,I)+kron(I,Z))
    else:
        Lm1 = eye(2**(L-1))
        Lm2 = eye(2**(L-2))
        return kron(Ham(L-1,ta,la),I)+kron(Lm2,kron(Z,Z))+ta*kron(Lm1,X)+la*kron(Lm1,Z) 
```

# Sampling

## Coin-toss like sampling

The ground state is a (complex in general, but in our case it is actually a purely real) vector of $2^L$ components:

$$|g.s.\rangle \equiv |g_1,g_2,...,g_{2^L}\rangle$$

where $g_i$, as said (as guaranteed by the Perron-Frobenius teorem) is a purely real number, thus

$$g_i^*g_i = g_i^2$$

where $z^*$ stands for complex conjugate of $z$. And so if one makes the following array:

$$\mathbf{p}^{g.s} \equiv |g_1^2,g_2^2,\dots,g_{2^L}^2\rangle$$

then $\mathbf{p}^{g.s}_i$ is the **probability** of observing $i$-th "bitstring" (= configuration of $\sigma_z$ observables of magnetizations on sites) if the system is prepared in the ground state.

The task is now to sample such bitstrings with corresponding probabilities. If one has access to the ground state wavefunction (the $|g.s.\rangle$ vector) – then one simple way to do this is such: consider the interval $(0,1)$

partitioned into intervals with respect to the probabilities 

$$(p_1,p_2,\dots,p_{2^L}) = (g_1^2,g_2^2,\dots,g_{2^L}^2)$$

then a "coin" is tossed – that is, a uniformly-distributed (over $(0,1)$) random number is generated, and the string that corresponds to the interval where the "coin" fell is thus sampled. On the image above, $L=2$ – there are two spins, thus there are four basis bitstrings: $\{s_1 = 00,s_2 = 01,s_3 = 10,s_4 = 11\}$ and the "coin" fell on the $01$ interval, thus the $01$ string is returned.

Such sampling can be done with one `numpy` package function – `numpy.random.choice`!

Generate a non-uniform random sample from `np.arange(4)` of size 10:

```python
np.random.choice(5, 10, p=[0.1, 0, 0.3, 0.6, 0])
```

To get the 'bitstrings' from the sampled integers ony simply has to convert them to binary.

```python
bin(3)
```

# d (g.s. distribution) / d lambda

Let us now do the following: we know that the ground-state vector squared of our Hamiltonian at a certain value of the perturbative parameter, $H(\lambda)$, is a probability distribution on the set of bitstrings: $p_\lambda(\textbf{s})$

For two close values of this perturbative parameter, $\lambda$ and $\lambda+\delta\lambda$, one can obtain the corresponding (ground-state) probability distributions and measure their **KL-divergence**, for example:

$$KL[p\,||\,q] = \sum_{\textrm{all bitstrings }\mathbf{s}} 
p(\mathbf{s})\,\log\left(\frac{p(\mathbf{s})}{q(\mathbf{s})}\right)$$

which is actually a very physical quantity, since $KL[p\,||\,q] = H(p,q)-H(p)$, where $H(p,q)$ is the (symmetric) relative entropy, and $H(p)$ – just the entropy.

What we'd like to do is the following: let's take the discredised interval (grid) of $\lambda$ values and measure the KL-divergences of the g.s. probability distributions along this interval:

$$KL[p(\lambda)\,||\,p(\lambda+\delta\lambda)]$$

as a function of $\lambda$

when choosing the interval of $\lambda$ values, keep in mind that the estimate of critical $\lambda_c$ value (where something should go wrong) is the typical energy level spacing 

$$\lambda_c \approx \frac{2L}{2^L} = L\cdot2^{1-L}$$

so, for 8 spins:

```python
L = 8
print(L*2**(1-L))
```

```python
la_grid = np.linspace(0.,0.1,40)
print(la_grid)
```

```python
def KL(p, q):
    # p and q should be same (proper) size numpy arrays
    return (p*np.log(p/q)).sum()

def Hrel(p, q):
    # the relative entropy
    return (p*np.log(q)).sum()
```

```python
p = np.array([1./3,1./3,1./3])
q1 = np.array([1./3,1./3,1./3])
q2 = np.array([1./4,1./6,0.])
```

```python
print(KL(p,q1))
print(Hrel(p,q2))
```

As you can see, KL divergence has troubles when some of the $p$ (or $q$) values turns zero, so one should maybe treat such cases separately (for example, changing $0.$ to some very small machine-epsilon number). 

But it turns out for our model that even though the ground state vector is quite localized in the 'bitstring-basis', none of its components is precisely zero:

```python
L = 5

H = Ham(L = L, ta = 1., la = 0.05)
Es, Vs = las.eigsh(H, k=2)
gs = Vs[:,0]

print(min(gs**2))
print(np.log(min(gs**2)))
```

And we can proceed with the task of interest – plotting $KL[p(\lambda)\,||\,p(\lambda+\delta\lambda)]$

```python
KL(p,q1)
```

```python
L = 9

# grid of lambda parameter values, N = 40 points
la_grid = np.linspace(0.,0.75,100)
# list of ground-state vectors
gs_s = []
# list of ground-state energies
gs_Es = []
# list of KL divergences, of size (N-1) values
KLs = []

# initializing the lists with the zeroth (lambda = 0) step
H = Ham(L = L, ta = 1., la = 0.)
Es, Vs = las.eigsh(H, k=2)
gs_s.append(Vs[:,0])
gs_Es.append(Es[0])

for la in la_grid[1:]:
    H = Ham(L = L, ta = 1., la = la)
    Es, Vs = las.eigsh(H, k=10)
    # compute the KL divergence with the previous (last element in gs_s) vector
    #KLs.append(KL(gs_s[-1]**2, Vs[:,0]**2))
    KLs.append(KL(gs_s[0]**2, Vs[:,0]**2))
    # append the current ground state vector and energy to the lists
    gs_s.append(Vs[:,0])
    gs_Es.append(Es[0])
    #print 'step lambda = ', la, ' done'
```

```python
plt.plot(la_grid[1:],KLs,'.')
plt.show()
```

## The full-eigenbasis-overlap matrix

Another plot that we'd like to look at to reveal system's behaviour is as follows.

So there is a grid of $N_\lambda$ discrete $\lambda$ values for which we study (fully diagonalize) the model's Hamiltonian. Denote a certain value from this grid by $\lambda_j$ and the corresponding eigenbasis by $\{|E^{(j)}_i\rangle\}$

Such eigenbases fully describe the model's physics, actually. So in some sense, if the two eigenbases are "similar" – then the physics of these two cases is similar. Let's describe how we'll be measuring this similarity.

For two different $\lambda$ values from the grid, $\lambda_{i}$ and $\lambda_{j}$, construct the following matrix $M_{nk}$ whose matrix elements are

$$M^{(ij)}_{nk} = \langle E^{(i)}_n | E^{(j)}_k \rangle$$

so if $i=j$, this is simply a Gram matrix of the $i^{th}$ $\lambda$ value eigenbasis – which is simply an identity matrix, since $|E^{(j)}_i\rangle$ – outputs of the diagonalization algorithm – are an ortho-normal set of vectors.

If $\lambda_i$ and $\lambda_j$ are close values, then $M^{(ij)}$ should be close to identity matrix. Thus $M^{(ij)}$-s "difference" from the identity matrix is a measure of eigenbases dissimilarity.

So now for a cartesian product of the two $\lambda$ value grids, 

$$(0,\Delta\lambda,2\Delta\lambda,\dots,2\lambda_c)\times(0,\Delta\lambda,2\Delta\lambda,\dots,2\lambda_c)$$

we can make a 3D (or a colored contour 2D) plot of

$$||M^{(ij)}-I||_2$$

the difference norm, where, say, the (x,y) coordinates (in case of 3D plot) are $(\lambda_i,\lambda_j)$, and the z-coordinate is this difference norm.

The choice of matrix norm is not so important really, and $L_2$ norm is the most usual one.

```python
L = 7

# grid of lambda parameter values, N = 40 points
la_grid = np.linspace(0.,0.75,40)
# list of ground-state vectors
Vs_s = []

for lam in la_grid:
    H = Ham(L = L, ta = 1., la = lam).toarray()
    ## WARNING!!! I'M BROADCASTING THE MATRIX TO A DENSE ONE HERE 
    ## TO GET THE FULL EIGENBASIS! THIS IS NOT EFFECTIVE FOR LARGE SIZES
    _, Vs = np.linalg.eigh(H)
    # compute the KL divergence with the previous (last element in gs_s) vector
    #KLs.append(KL(gs_s[-1]**2, Vs[:,0]**2))
    Vs_s.append(Vs)
```

Here's the $\langle E^{(0)}_n | E^{(0)}_k \rangle$ overlap matrix – that is, the Gram matrix of $\lambda_0 = 0$ eigenbasis:

```python
plt.imshow(np.abs(Vs_s[0].T.dot(Vs_s[0])))
plt.xlabel('$k$ of $| E^{(0)}_k >$')
plt.ylabel('$n$ of $< E^{(0)}_n |$')
plt.title('$<E^{(0)}_n | E^{(0)}_k>$')
plt.colorbar()
plt.show()
```

Here's the overlap matrix of $\lambda_0=0$ and $\lambda_2\approx0.04$:

```python
la_grid[2]
```

```python
plt.imshow(np.abs(Vs_s[0].T.dot(Vs_s[2])))
plt.xlabel('$k$ of $| E^{(2)}_k >$')
plt.ylabel('$n$ of $< E^{(0)}_n |$')
plt.title('$<E^{(0)}_n | E^{(2)}_k>$')
plt.colorbar()
plt.show()
```

So one can see that **the low-energy eigenstates remain very close as $\lambda$ smoothly grows, while the high-energy eigenstates are wildly mixing**

And here's the overlap matrix of $\lambda_0=0$ and $\lambda_{35}\approx0.67$:

```python
plt.imshow(np.abs(Vs_s[0].T.dot(Vs_s[35])))
plt.xlabel('$k$ of $| E^{(35)}_k >$')
plt.ylabel('$n$ of $< E^{(0)}_n |$')
plt.title('$<E^{(0)}_n | E^{(35)}_k>$')
plt.colorbar()
plt.show()
```

Now let's compute the norms of $$||M^{(ij)}-I||_2$$ over the $40\times 40$ grid (Cartesian product) of $(\lambda_i,\lambda_j)$ and see what we get:

```python
Overlaps = []

for i in range(40):
    OverlapsRow = []
    for j in range(40):
        OverlapsRow.append(np.linalg.norm(Vs_s[i].T.dot(Vs_s[j])))
    Overlaps.append(OverlapsRow)

Overlaps = np.array(Overlaps)
```

```python
plt.imshow(Overlaps)
plt.xlabel('$j$ of $\lambda_j$, 40 points')
plt.ylabel('$i$ of $\lambda_i$, 40 points')
plt.title('$||M^{(ij)}-I||_2$')
plt.colorbar()
plt.show()
```

And, due to the above mentioned wild mixing of high-energy eigenstates, this clearly has little sense for the full eigenbasis overlap matrices.

## Ground-states overlap as function of lambda

Apart from the above KL-divergence test, another (pretty much the same, only without the squares to get the distribution from the state vector) way to look at the transition is to see how much the ground states (and let's also look at several excited states above the ground state) change with growth of $\lambda$. 

How much they differ = how far their scalar product is from 1.

Here's a plot of ground state vectors' $2^7 = 128$ components for several $\lambda$ values:

```python
plt.plot(np.real(Vs_s[0][:,0]),'-.',label = '$\lambda = $'+str(round(la_grid[0],2)))
plt.plot(np.real(Vs_s[2][:,0]),'-.',label = '$\lambda = $'+str(round(la_grid[2],2)))
plt.plot(np.real(Vs_s[15][:,0]),'-.',label = '$\lambda = $'+str(round(la_grid[15],2)))
plt.plot(np.real(Vs_s[25][:,0]),'-.',label = '$\lambda = $'+str(round(la_grid[35],2)))
plt.legend()
plt.show()
```

So the structure of the g.s. vector changes quite a lot.

Let's see the overlap dependence on $\lambda$. 

We will plot it for g.s. as well as for several "excited" states – `g.s. + 1` means "next level above the ground state". Hope its clear.

```python
gs0 = Vs_s[0][:,0]

plt.plot(la_grid[1:],[np.abs(Vs_s[0][:,0].dot(Vs_s[i][:,0])) for i in range(1,40)],'--',label='vec = g.s.')
plt.plot(la_grid[1:],[np.abs(Vs_s[0][:,1].dot(Vs_s[i][:,1])) for i in range(1,40)],'--',label='vec = g.s. + 1')
plt.plot(la_grid[1:],[np.abs(Vs_s[0][:,2].dot(Vs_s[i][:,2])) for i in range(1,40)],'--',label='vec = g.s. + 2')
plt.plot(la_grid[1:],[np.abs(Vs_s[0][:,3].dot(Vs_s[i][:,3])) for i in range(1,40)],'--',label='vec = g.s. + 3')

plt.title('$|<vec^{(0)}|vec^{(\lambda)}>|$')
plt.xlabel('$\lambda$')
plt.legend()
plt.show()
```

So, clearly, **the more excited the state – the more it is mixing with other close eigenstates**

# Training an RBM to reproduce ground-state Z-distribution

Let us now prepare the procedure of sampling from the ground-state to have a large dataset to train the RBM on.

```python
from sklearn.neural_network import BernoulliRBM
```

```python
L = 6

H = Ham(L = L, ta = 1., la = 0.1)
Es, Vs = las.eigsh(H, k=2)
gs = Vs[:,0]
```

```python
X_raw = np.random.choice(2**L, 10000, p=gs**2)
```

```python
plt.bar(range(2**L),gs**2,color='green',alpha=0.75,label='True $p(s)$')
plt.hist(X_raw,color='red',alpha=0.5,density=True,label='Sampled',bins=2**L)
plt.legend()
plt.title('Sampled 10.000 bitstrings')
plt.show()
```

Now remember that `X_raw` is a 1D-array of numbers that we need to reshape to:

$$(N_{samples},\,N_{observables})$$ 

so to make a 2D array, rows being bitstrings.

```python
def bitli(L,n):
    li = [int(el) for el in list("{0:b}".format(n))]
    return [0]*(L-len(li)) + li

def inv_bitli(li):
    return int(''.join(map(str,li)), 2)
```

and we'll also need the inverse operation:

```python
n_try = 5
print(n_try)
bin_li_try = bitli(3, n_try)
print(bin_li_try)
n_back = inv_bitli(bin_li_try)
print(n_back)
```

```python
X = np.array([bitli(L,n) for n in X_raw])
```

```python
X.shape
```

```python
model = BernoulliRBM(n_components=3, learning_rate=0.1, batch_size=100, n_iter=100, verbose=0)
```

```python
model.fit(X[:5000,:])
```

```python
X_new = model.gibbs(X[5000:,:]).astype(int)
```

```python
X_new_raw = [inv_bitli(list(el)) for el in list(X_new)]
```

```python
plt.bar(range(2**L),gs**2,color='green',alpha=0.75,label='True $p(s)$')
plt.hist(X_raw,color='red',alpha=0.5,density=True,label='Sampled',bins=2**L)
plt.hist(X_new_raw,color='blue',alpha=0.5,density=True,label='RBM-Sampled',bins=2**L)
plt.title('Sampled 100.000 bitstrings')
plt.show()
```
