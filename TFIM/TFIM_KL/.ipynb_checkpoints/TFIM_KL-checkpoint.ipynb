```python
import numpy as np
from scipy.sparse import csr_matrix, kron, eye
import scipy.sparse.linalg as las
import scipy.linalg as la
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
I = csr_matrix(np.array([[1, 0], 
                         [0, 1]]))
X = csr_matrix(np.array([[0, 1], 
                         [1, 0]]))
Y = csr_matrix(np.array([[0, -1j], 
                         [1j, 0]]))
Z = csr_matrix(np.array([[1, 0], 
                         [0, -1]]))
```

```python
# setting up main parameters of the model


# number of steps in the grid of longitudinal field values
D  = 1000
# grid of transverse field values
ta = np.linspace(0,2,D)
```

```python
# this function returns a quantum many-body Hamiltonian of the transverse field Ising model
def Ham(L=3, ta = 0.):
    if L == 2:
        return - kron(Z,Z) - ta * ( kron(X,I) + kron(I,X) )
    else:
        Lm1 = eye(2**(L-1))
        Lm2 = eye(2**(L-2))
        return kron(Ham(L-1,ta),I) - kron(Lm2,kron(Z,Z)) - ta * kron(Lm1,X)
```

```python
# define KL-divergence between two probability distributions, p and q should be same size numpy arrays
def KL(p, q):
    return (p*np.log(p/q)).sum()
```

```python
%%time
for L in np.arange(7,20):
    KLs=[]
    for i in range(D-1):
        KLs.append(KL(las.eigsh(Ham(L,ta[i]), k=10)[1][:,0]**2,las.eigsh(Ham(L,ta[i+1]), k=10)[1][:,0]**2))
    np.save('Kls/'+str(L)+'_KLs.npy',KLs)
```

```python
plt.figure(figsize=(9,4.5))
for L in np.arange(7,20):
    plt.plot(ta[:-1],np.load(str(L)+'_KLs.npy'))
plt.xlabel('$τ$')
plt.ylabel('$D$',rotation=0,labelpad=10)
plt.legend(['L=7','L=8','L=9','L=10','L=11','L=12','L=13','L=14','L=15','L=16','L=17','L=18','L=19','L=20'])
plt.ylim((0.000,0.0001))
plt.xlim((0.4,1.8))
plt.xticks(np.arange(0.4,2.0,0.2))
plt.tight_layout()
plt.savefig('TFIM_KL_Scaling.pdf',transparent=True)
```
