```python
import numpy as np
from scipy.sparse import csr_matrix, kron, eye
import scipy.sparse.linalg as las
import scipy.linalg as la
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
%matplotlib inline
```

```python
font = {'size'   : 14}
plt.rc('font', **font)
```

```python
I = csr_matrix(np.array([[1, 0], 
                         [0, 1]]))
X = csr_matrix(np.array([[0, 1], 
                         [1, 0]]))
Y = csr_matrix(np.array([[0, -1j], 
                         [1j, 0]]))
Z = csr_matrix(np.array([[1, 0], 
                         [0, -1]]))
```

```python
# setting up main parameters of the model

# number of steps in the grid of longitudinal field values
D  = 40
# grid of transverse field values
ta = np.linspace(0,2,D)
```

```python
# this function returns a quantum many-body Hamiltonian of the transverse field Ising model
def Ham(L=3, ta = 0.):
    if L == 2:
        return - kron(Z,Z) - ta * ( kron(X,I) + kron(I,X) )
    else:
        Lm1 = eye(2**(L-1))
        Lm2 = eye(2**(L-2))
        return kron(Ham(L-1,ta),I) - kron(Lm2,kron(Z,Z)) - ta * kron(Lm1,X)
```

```python
%%time
for L in [21]:
    
    dEs = []
    for i in range(len(ta)):
        Es = las.eigsh(Ham(L,ta[i]),k=10)[0]
        dE = Es[1] - Es[0]
        dEs.append(dE)
    
    np.save('Gaps/'+str(L)+'Gaps.npy',dEs)
```

```python
plt.figure(figsize=(9,4.5))
for L in np.arange(7,21):
    plt.plot(ta,np.load(str(L)+'Gaps.npy'))
plt.legend(['L=7','L=8','L=9','L=10','L=11','L=12','L=13','L=14','L=15','L=16','L=17','L=18','L=19','L=20'],fontsize=11)
plt.xlabel('$τ$')
plt.xlim((0,2))
plt.ylim((0,2.2))
plt.yticks(np.arange(0,2.4,0.2))
plt.xticks(np.arange(0,2.2,0.2))
plt.ylabel('$\Delta$', rotation=0, labelpad=10)
plt.tight_layout()
plt.savefig('TFIM_Gap_Scaling.pdf', transparent=True)
```

```python
def fitfunc_1(ta,ta_c,gamma):
    return abs(ta-ta_c)**gamma
```
